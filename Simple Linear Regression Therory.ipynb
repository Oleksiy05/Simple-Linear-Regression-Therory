## <center>Simple Linear Regression Therory</center>

Simple Linear Regression models shows a linear relation between a dependent variable y and independent variable x. It has a simple equation, of degree 1, for example $$y = 2*(x) + 3$$

### Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5.0, 5.0, 0.1)

##We can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise

plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```
